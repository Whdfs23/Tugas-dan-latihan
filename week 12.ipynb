```python

# Write your code below and press Shift+Enter to execute

def hitung_luas_persegi():
    try:
        sisi = input("Masukkan panjang sisi persegi: ")
        if not sisi:
            raise ValueError("tidak boleh kosong")
        if not sisi.isdigit():
            raise ValueError("hanya boleh memasukkan angka")
        sisi = int(sisi)
        luas = sisi * sisi
        print("Luas persegi:", luas)
    except ValueError as e:
        print(e)

def hitung_luas_lingkaran():
    try:
        jari_jari = input("Masukkan jari-jari lingkaran: ")
        if not jari_jari:
            raise ValueError("tidak boleh kosong")
        if not jari_jari.isdigit():
            raise ValueError("hanya boleh memasukkan angka")
        jari_jari = int(jari_jari)
        luas = 3.14 * jari_jari * jari_jari
        print("Luas lingkaran:", luas)
    except ValueError as e:
        print(e)

def hitung_luas_persegi_panjang():
    try:
        panjang = input("Masukkan panjang: ")
        if not panjang:
            raise ValueError("tidak boleh kosong")
        if not panjang.isdigit():
            raise ValueError("hanya boleh memasukkan angka")
        panjang = int(panjang)
        lebar = input("Masukkan lebar persegi panjang: ")
        if not lebar:
            raise ValueError("tidak boleh kosong")
        if not lebar.isdigit():
            raise ValueError("hanya boleh memasukkan angka")
        lebar = int(lebar)
        luas = panjang * lebar
        print("Luas persegi panjang:", luas)
    except ValueError as e:
        print(e)

while True:
    print("\nMenu:")
    print("1. Hitung Luas Persegi")
    print("2. Hitung Luas Lingkaran")
    print("3. Hitung Luas Persegi Panjang")
    print("4. Keluar")

    pilihan = input("Masukkan pilihan Anda: ")

    if pilihan == '1':
        hitung_luas_persegi()
    elif pilihan == '2':
        hitung_luas_lingkaran()
    elif pilihan == '3':
        hitung_luas_persegi_panjang()
    elif pilihan == '4':
        break
    else:
        print("Pilihan tidak valid.")
```

```python
NPM = input("Masukkan NPM anda :")

try:
    jumlah_genap = 0
    jumlah_ganjil = 0
    for angka in NPM:
        if not angka.isdigit():
            raise ValueError("NPM harus berisi angka saja")
        angka_int = int(angka)
        if angka_int % 2 == 0:
            jumlah_genap += 1
        else:
            jumlah_ganjil += 1
    print("Jumlah genap :", jumlah_genap)
    print("Jumlah ganjil :", jumlah_ganjil)

except ValueError as S:
    print(f"Error: {S}")
except Exception as S:
    print(f"Terjadi kesalahan: {S}")
```
